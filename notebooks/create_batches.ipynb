```python
import pandas as pd
pd.set_option('max_colwidth', 1000)
import numpy as np
import os
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
source = r'D:\ML_projects\IPV-Scraper\results\scraped_tweets_cleaned_22-04-022.csv'
dest = r'D:\ML_projects\IPV-Scraper\results\batches\twitter'
```

```python
df = pd.read_csv(source, encoding='utf-8', skip_blank_lines=True)
df.sample(10)
```

```python
tweets = df['tweet'].sample(frac = 1.0, random_state=100)
tweets.shape
```

```python
# Add sentence enders.
enders = ['|', '?', '।', '.', '!']
tweets = tweets.apply(lambda x : x + " |" if x[-1] not in enders else x)

tweets
```

```python
tweets = tweets.astype("str")
tweets.sample(10)
```

```python
for k, ii in enumerate(range(0, len(tweets) + 1, 10), 1):
    start = ii
    end = ii + 10
    batch = tweets.iloc[start:end].str.strip()

    # Filename.
    filename = f'batch_twitter_{start}-{end}.txt'
    
    if k%2 == 0:
        batch.to_csv(os.path.join(dest, 'shr', filename), header = None, index = None, encoding = 'utf-8')
    else:
        batch.to_csv(os.path.join(dest, 'krn', filename), header = None, index = None, encoding = 'utf-8')
```
