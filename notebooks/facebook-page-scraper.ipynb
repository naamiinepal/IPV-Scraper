# Scrape facebook posts' content from public page using username.

```python
import os
from pprint import pprint
from facebook_page_scraper import Facebook_scraper
```

```python
page_name = "officialroutineofnepalbanda"
posts_count = 10
browser = "firefox"
timeout = 600 #600 seconds

meta_ai = Facebook_scraper(page_name, posts_count, browser, timeout=timeout)
```

## Save to JSON.

```python
json_data = meta_ai.scrap_to_json()
pprint(json_data)
```

## Save to CSV.

```python
filename = "scraped_ronb"  #file name without CSV extension,where data will be saved
directory = r"D:\ML_projects\IPV-Scraper\results\fb-scrape" #directory where CSV file will be saved
meta_ai.scrap_to_csv(filename,directory)
```
